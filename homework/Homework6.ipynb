```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#1
def sinusoid(x, A, w):
    return A * np.sin(w * x)
```

```python
#1.2
x = np.linspace(0, 2*np.pi, 1000)
```

```python
#1.3
A = np.linspace(0, 10, 5)
w = np.linspace(0, 10, 5)
```

```python
#1.4
plt.figure()
```

```python
#1.5
y = np.array([])
for i in range(5):
    calculate = sinusoid(x, A[i], w[i])
    y = np.append(y, calculate)
plt.plot(y)
```

```python
#2
one = []
n = 40
for i in range(n):
    one.append(random.randint(0,101))
two = []
n = 40
for i in range(n):
    two.append(random.randint(0,101))
```

```python
plt.figure(figsize = (10, 10))
plt.plot(one, color = "orange", linewidth = 10)
plt.plot(two, color = "red", linestyle = "--")
plt.xlabel("x data")
plt.ylabel("y data")
plt.legend(["orange", "red"])
plt.title("graph")
```
